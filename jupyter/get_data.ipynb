```python
import os 
import pandas as pd
import glob 
import re 
from tqdm.notebook import tqdm
import warnings
import os.path
#import datetime
```

```python
url_data = "https://www.mscbs.gob.es/profesionales/saludPublica/ccayes/alertasActual/nCov/documentos/Informe_Comunicacion_20210122.ods"
```

```python
df_external = pd.read_excel(url_data, engine="odf")
columns = df_external.columns.values
columns[0] = 'comunidades'
df_external.columns = columns
df_external['Fecha de la última vacuna registrada (2)']= df_external['Fecha de la última vacuna registrada (2)'].dt.normalize()
df_external.reset_index(drop=True, inplace=True)

```

```python
# Merge daily and total data
file="../data/vaccines_spain_total.csv"
if os.path.isfile(file):
    df_historic = pd.read_csv(file)
    df_historic['Fecha de la última vacuna registrada (2)']= df_external['Fecha de la última vacuna registrada (2)'].dt.normalize()
    df_historic.reset_index(drop=True, inplace=True)
    df = pd.concat([df_external,df_historic])
    # Remove duplicates
    df.reset_index(drop=True, inplace=True)
    df = df.drop_duplicates(subset=['comunidades','Fecha de la última vacuna registrada (2)'])
    

    df.to_csv(file, index=False)
else:
    df_external.to_csv(file, index=False)
```
